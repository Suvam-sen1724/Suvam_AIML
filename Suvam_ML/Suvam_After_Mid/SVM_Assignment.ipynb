Question 1: Reading and Exploring the Iris Dataset
1. Read the iris dataset stored in .data format from the location Data and
store it as a DataFrame.
2. Print the first five rows of the DataFrame to examine the dataset.
3. Display the statistical insights of the dataset using the describe() function.
4. Visualize the relationship between the ’sepal-length’ and ’class’ using a
boxplot.
5. Visualize the relationship between the ’sepal-width’ and ’class’ using a
boxplot.
6. Plot a bar graph representing the distribution of classes in the dataset.

```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Read the iris dataset stored in .data format from the location Data and store it as a DataFrame.
df = pd.read_csv('iris.data', header=None)

# Assign column names
df.columns = ['sepal-length', 'sepal-width', 'petal-length', 'petal-width', 'class']

# 2. Print the first five rows of the DataFrame to examine the dataset.
print(df.head())

# 3. Display the statistical insights of the dataset using the describe() function.
print(df.describe())

# 4. Visualize the relationship between the ’sepal-length’ and ’class’ using a boxplot.
plt.figure(figsize=(10, 7))
sns.boxplot(x='class', y='sepal-length', data=df)
plt.title('Boxplot of Sepal Length by Class')
plt.show()

# 5. Visualize the relationship between the ’sepal-width’ and ’class’ using a boxplot.
plt.figure(figsize=(10, 7))
sns.boxplot(x='class', y='sepal-width', data=df)
plt.title('Boxplot of Sepal Width by Class')
plt.show()

# 6. Plot a bar graph representing the distribution of classes in the dataset.
plt.figure(figsize=(10, 7))
sns.countplot(x='class', data=df)
plt.title('Distribution of Classes in the Dataset')
plt.show()
```

Question 2: SVM Model with Linear Kernel
1. Separate the independent variable (features) and dependent variable (target)
into two separate variables, X and y respectively.
2. Split the dataset into training and testing sets with a test size of 33% and
a random state of 42.
3. Build an SVM model using a linear kernel.
4. Fit the SVM model to the training data.
5. Predict the target variable using the fitted model.
6. Print the confusion matrix to evaluate the performance of the model.

```python
# Import necessary libraries
from sklearn.model_selection import train_test_split
from sklearn import svm
from sklearn.metrics import confusion_matrix

# 1. Separate the independent variable (features) and dependent variable (target) into two separate variables, X and y respectively.
X = df[['sepal-length', 'sepal-width', 'petal-length', 'petal-width']]
y = df['class']

# 2. Split the dataset into training and testing sets with a test size of 33% and a random state of 42.
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)

# 3. Build an SVM model using a linear kernel.
model = svm.SVC(kernel='linear')

# 4. Fit the SVM model to the training data.
model.fit(X_train, y_train)

# 5. Predict the target variable using the fitted model.
y_pred = model.predict(X_test)

# 6. Print the confusion matrix to evaluate the performance of the model.
print(confusion_matrix(y_test, y_pred))
```

Question 3: SVM Model with Polynomial Kernel
1. Build an SVM model using a polynomial kernel.
2. Fit the SVM model to the training data.
3. Predict the target variable using the fitted model.
4. Print the confusion matrix to evaluate the performance of the model.

```python
# Import necessary libraries
from sklearn import svm
from sklearn.metrics import confusion_matrix

# 1. Build an SVM model using a polynomial kernel.
model_poly = svm.SVC(kernel='poly')

# 2. Fit the SVM model to the training data.
model_poly.fit(X_train, y_train)

# 3. Predict the target variable using the fitted model.
y_pred_poly = model_poly.predict(X_test)

# 4. Print the confusion matrix to evaluate the performance of the model.
print(confusion_matrix(y_test, y_pred_poly))
```

Question 4: SVM Model with Radial Kernel
1. Build an SVM model using a radial kernel.
2. Fit the SVM model to the training data.
3. Predict the target variable using the fitted model.
4. Print the confusion matrix to evaluate the performance of the model.

```python
# Import necessary libraries
from sklearn import svm
from sklearn.metrics import confusion_matrix

# 1. Build an SVM model using a radial kernel.
model_radial = svm.SVC(kernel='rbf')

# 2. Fit the SVM model to the training data.
model_radial.fit(X_train, y_train)

# 3. Predict the target variable using the fitted model.
y_pred_radial = model_radial.predict(X_test)

# 4. Print the confusion matrix to evaluate the performance of the model.
print(confusion_matrix(y_test, y_pred_radial))
```

Question 5: Comparison of SVM Models
Compare the accuracies of the SVM models built with linear, polynomial, and
radial kernels.

```python
# Import necessary libraries
from sklearn.metrics import accuracy_score

# Predict the target variable using the fitted models
y_pred_linear = model.predict(X_test)
y_pred_poly = model_poly.predict(X_test)
y_pred_radial = model_radial.predict(X_test)

# Calculate the accuracies of the models
accuracy_linear = accuracy_score(y_test, y_pred_linear)
accuracy_poly = accuracy_score(y_test, y_pred_poly)
accuracy_radial = accuracy_score(y_test, y_pred_radial)

# Print the accuracies
print(f"Accuracy of SVM model with linear kernel: {accuracy_linear}")
print(f"Accuracy of SVM model with polynomial kernel: {accuracy_poly}")
print(f"Accuracy of SVM model with radial kernel: {accuracy_radial}")
```
